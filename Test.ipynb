```python
required_columns = [
    "Church_Affiliation",
    "First_Name",
    "Last_Name",
    "Birthday",
    "Home_Address"
]

# Drop rows with missing/NaN/empty values in the required columns
df_filled = df_test.dropna(subset=required_columns)

# Remove trimmed empty strings too
for col in required_columns:
    df_filled = df_filled[df_filled[col].astype(str).str.strip() != '']

# Keep just "Is_Member" and "Opt_In_Announcements" in addition (no missing enforcement for these)
cols_to_show = required_columns + ["Is_Member", "Opt_In_Announcements", "Wedding_Anniversary"]
df_filled = df_filled[cols_to_show].reset_index(drop=True)
df_filled
```

```python
# Summary statistics for numeric/bool columns
desc_num = df_filled.describe(include=[float, int, bool])
for col in desc_num.columns:
    print(f"For numeric/bool column '{col}':")
    for stat in desc_num.index:
        print(f"    {stat.capitalize()}: {desc_num.at[stat, col]}")
    print()

# Summary statistics for object/categorical columns
desc_obj = df_filled.describe(include=[object])
for col in desc_obj.columns:
    print(f"For column '{col}':")
    count = desc_obj.at['count', col]
    unique = desc_obj.at['unique', col]
    top = desc_obj.at['top', col]
    freq = desc_obj.at['freq', col]
    print(f"    There are {int(count)} non-missing values.")
    print(f"    There are {int(unique)} unique values.")
    print(f"    The most common value is '{top}', occurring {int(freq)} times.")
    print()

# Show number of missing values per column
missing_vals = df_filled.isnull().sum()
for col, n_missing in missing_vals.items():
    if n_missing == 0:
        print(f"There are no missing values for column '{col}'.")
    else:
        print(f"There are {n_missing} missing values for column '{col}'.")

# Show count of unique values per column
unique_counts = df_filled.nunique()
for col, n_unique in unique_counts.items():
    print(f"Column '{col}' has {n_unique} unique value(s).")

# Show top 5 most frequent values for each column
for col in df_filled.columns:
    print(f"For column '{col}', the 5 most common values and their counts are:")
    vc = df_filled[col].value_counts().head()
    for val, count in vc.items():
        print(f"    '{val}': {count} times")
    print()
```
